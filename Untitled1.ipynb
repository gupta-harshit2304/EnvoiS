```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading a dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#printing first 5 rows of data set
credit_card_data.head()
```

```python
#printing last 5 rows of data set
credit_card_data.tail()
```

```python
#information of dataset
credit_card_data.info()
```

```python
#distribution of legit and fradudlent transactions
credit_card_data['Class'].value_counts()
```

This Dataset is highly Unbalanced.

0 --> Legit Transaction
1 --> Fraudlent Transaction

---



```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical analysis of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Comparing the values for both tranbsactions.
credit_card_data.groupby('Class').std()
```

```python
# Comparing the values for both tranbsactions.
credit_card_data.groupby('Class').mean()
```

Under-Sampling -(as the dataset is unbalanced)
Build a sample dataset containing similar distributions of legit and fraudlent transaction.

No. of Fraudlent transaction = 162


---



```python
legit_sample = legit.sample(n=162)
```

Now we will concatenate Two Data **Frames**

---



```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
#axis 0 means rowwise and axis 1 means columnwise
```

```python
#checking the results by counting the number of fraudlent and legit transaction
new_dataset['Class'].value_counts()
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.groupby('Class').std()
```

Splitting data into Features and Targets

---



```python
X = new_dataset.drop(columns = 'Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```



---


Split the data into training and testing data

---



```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```



---
Model Training - Logistic Regression
(As we are doing binary classification of data therefore LR will be the best choice)


---




```python
model = LogisticRegression()
```

```python
# Training the LR model with training data
model.fit(X_train,Y_train)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```



---

Model Evaluation


---



```python
# Accuracy on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("Accuracy on Training Data : ", training_data_accuracy)
```

```python
# Accuracy on Testing Data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("Accuracy on Testing Data : ", testing_data_accuracy)
```



---
As Both the accuracy are close to each other, which means our model is neither overfitting nor underfitting.
